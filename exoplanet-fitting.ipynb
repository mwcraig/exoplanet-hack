# Fitting exoplanet transit light curves

## Background

The Exopalnet Tansit Database  describes how they do fitting in [Poddaný, S., Brát, L., & Pejcha, O. (2010), New Astronomy, 15(3), 297–301. doi:10.1016/j.newast.2009.09.001](http://adsabs.harvard.edu/abs/2010NewA...15..297P).

Their fit is

$$
m(t_i) = A - 2.5\log F(z, p, c_2) + B (t-t_{mean}) + C (t-t_{mean})^2,
$$

where $F(z, p, c_2)$ gives the flux during the transit as a function of the projected separation of the planet and star, $z$. **Note: in this notebook the limb darkening coefficient that Poddany et al call $c_1$ is called $c_2$ for consistency with Mandel, et al.** 

The relative flux $F$ is from [Mandel, K., & Agol, E. (2002) The Astrophysical Journal, 580(2), L171–L175. doi:10.1086/345520](http://adsabs.harvard.edu/abs/2002ApJ...580L.171M), who consider several cases. 

ETD uses the case of "small planets," for which [Madel & Agol](http://adsabs.harvard.edu/abs/2002ApJ...580L.171M) have

$$
F = 
\begin{cases}
1,& z > 1+p,\\
1 - \frac{I^*(z)}{4\pi\Omega}\left[p^2\cos^{-1}\left(\frac{z-1}{p}\right) - (z-1)\sqrt{p^2 - (z-1)^2}\right], & 1-p < z < 1+p,\\
1 - \frac{I^*(z)}{4\Omega}p^2, & z \le 1-p,\\
\end{cases}
$$

where

$$
I^*(z) = 
\begin{cases}
(1-a)^{-1}\int_{z-p}^{1}I(r) 2rdr,& 1-p < z < 1+p,\\
\frac{1}{4zp} \int_{z-p}^{z+p}I(r) 2rdr,& z < 1-p
\end{cases}
$$

in which $a \equiv (z-p)^2$. **Note that there is a factor of $\pi$ missing from [Madel & Agol](http://adsabs.harvard.edu/abs/2002ApJ...580L.171M) Eq. (8) which is included in this notebook.**

The relative separation of  the planet and star is $z=d/r_*$, where $d$ is the separation of the centers of the star and the  planet and $r_*$ is the radius of the  star. The point of closest approach is $z_0=(a/r_*)\cos{i}$, where $a$ is the semi-major axis of the orbit and $i$ is the inclination of the orbit.

The parameter $p$ is the ratio of the planet's radius to the star's, $p=r_p/r_*$. 

One consequence of the small planet approximation is that its path across the star is a straight line in which the separation at closest approach is determined by the impact parameter $z_0$.


## Need $z(t)$

Note that [Madel & Agol](http://adsabs.harvard.edu/abs/2002ApJ...580L.171M) approach $z(t)$ differently; the approach below gives the same result as theirs, but is phrased in terms of planet radius $p$ and transit duration $D$ instead of orbital frequency and inclination. See the third paragraph of Section 6 for their formula.
 
Consider the chord shown below.

```python
from IPython.display import Image
Image('http://images.tutorvista.com/cms/images/67/length-of-chord.png')
```

In our case the impact parameter, labeled $d$, in the image above, is $z_0 r_*$, and the apprpriate radius is $r_* + r_p= r_*(1+p)$, because then the cord goes from the point of first contact of the planet with the disk of the star to the point of last contact. 

The length of the cord is then

$$
\ell = 2\sqrt{r_*^2(1+p)^2 - z_0^2 r_*^2} = 2 r_*\sqrt{(1+p)^2-z_0^2}.
$$

Assuming the planet's velocity is constant while moving across the face of the disk, that velocity is $v=\ell/D$ so that at time $t$ the planet is a distance 

$$
\delta = \frac{\ell}{D} (t -t_0)
$$ 

from the center point. The apparent separation is then

$$
d_{sep} = \sqrt{\delta^2 + (z_0 r_*)^2} = r_* \sqrt{(\delta/r_*)^2 + z_0^2},
$$

and

$$
\begin{align}
\delta^2 &= \frac{(t-t_0)^2}{D^2} \ell^2 \\
& = \frac{(t-t_0)^2}{D^2} 4 r_*^2 ((1+p)^2-z_0^2) \\
& = 4r_*^2 \frac{(t-t_0)^2}{D^2}  ((1+p)^2-z_0^2).
\end{align}
$$

Putting this back into the expression for $d_{sep}$ gives

$$ 
d_{sep} = r_* \sqrt{\frac{4(t-t_0)^2}{D^2}((1+p)^2-z_0^2) + z_0^2} ,
$$

so that

$$
z = d_{sep}/r_* =  \sqrt{\frac{4(t-t_0)^2}{D^2}((1+p)^2 - z_0^2) + z_0^2}.
$$

As a check, [Madel & Agol](http://adsabs.harvard.edu/abs/2002ApJ...580L.171M) state, in the paragraph just above their Eq. (8) in the section about the small planet approximation, that the phase at 1/4 or 3/4 through the transit is $(1+3z_0^2)^{1/2}/2$. The formula above, evaluated at $t=t_0 \pm D/4$, gives

$$
\begin{align}
z &= \sqrt{\frac{4D^2/16}{D^2}((1+p)^2 - z^2_0) + z^2_0} \\
& = \sqrt{\frac{(1+p)^2-z_0^2}{4} + z_0^2} \\
& = \sqrt{\frac{(1+p^2)+3z_0^2}{4}} \\
& = \frac{\sqrt{1+3z_0^2}}{2} \sqrt{1 + \frac{2p+p^2}{1+3z_0^2}}\\
& \approx \frac{\sqrt{1+3z_0^2}}{2},
\end{align}
$$

if $p \ll 1$, which agrees with [Madel & Agol](http://adsabs.harvard.edu/abs/2002ApJ...580L.171M).

## Limb darkening

### Case 1: $1-z < p < 1+z$

[Poddaný et al (2010)](http://adsabs.harvard.edu/abs/2010NewA...15..297P) model limb darkening with "the linear law with coefficient $c_1$," by which I believe they mean

$$
I(r) = 1 - c_2 (1-\mu),
$$

with $\mu=(1-r^2)^{1/2}$, for $0 \le r \le 1$. **Note that what [Madel & Agol](http://adsabs.harvard.edu/abs/2002ApJ...580L.171M) and this notebook call $c_2$ is what [Poddaný et al (2010)](http://adsabs.harvard.edu/abs/2010NewA...15..297P) call  $c_1$.** They end up assuming $c_2=0.5$ in all of their cases because trying to fit $c_2$ along with other parameters led to bad fits.

This enters the function $F(z, p, c_2)$ through the integral

$$
I^*(z) = (1-a)^{-1}\int_{z-p}^{1}I(r) 2rdr,
$$

which becomes

$$
I^*(z) = \frac{2}{1-a} \int_{z-p}^1 \left(1 - c_2 \left(1-(1-r^2)^{1/2}\right)\right) rdr
= \frac{2}{1-a} \left(\int_{z-p}^1 (1-c_2) r dr + c_2 \int_{z-p}^1 (1-r^2)^{1/2} r dr\right).
$$

The first part is trivial,

$$
\int_{z-p}^1 (1-c_2) r dr = (1-c_2) \left.\frac{r^2}{2}\right|_{z-p}^1 = (1-c_2)(1-(z-p)^2)/2,
$$

while the second part is only a little less trivial if you make the substitution $u=1-r^2$ (so that $du=-2rdr$)

$$
\int_{z-p}^1 (1-r^2)^{1/2} r dr = -\frac{1}{2}\int_{1-(z-p)^2}^{0} u^{1/2} du
= -\frac{1}{2}\frac{2}{3} \left.(1-r^2)^{3/2}\right|_{(z-p)}^1
= \frac{1}{3}(1-(z-p)^2)^{3/2}.
$$

Putting both of these results into the expression above for $I^*(z)$ gives

\begin{align}
I^*(z) &= \frac{2}{1-a} \left(\frac{1-c_2}{2}(1-(z-p)^2) + \frac{c_2 }{3}(1-(z-p)^2)^{3/2}\right)\\
&= \frac{1-(z-p)^2}{1-a} \left(1-c_2 + \frac{2c_2}{3}\sqrt{(1-(z-p)^2)}\right) \\
&= 1-c_2 + \frac{2c_2}{3}\sqrt{1-(z-p)^2},
\end{align}

since $a \equiv (z-p)^2$. Messy, but not impossible to work with.

### Case 2: $z<1-p$

In this case,

$$
I^*(z) = \frac{1}{4zp} \int_{z-p}^{z+p}I(r) 2rdr;
$$

as above the integral is really two pieces,

$$
\int I(r) rdr = \int \left(1 - c_2 \left(1-(1-r^2)^{1/2}\right)\right) rdr
= \left(\int (1-c_2) r dr + c_2 \int (1-r^2)^{1/2} r dr\right).
$$

From above (though with different limits of integration),

$$
\begin{align}
\int_{z-p}^{z+p} (1-c_2) r dr &= (1-c_2) \left.\frac{r^2}{2}\right|_{z-p}^{z+p} \\
& = \frac{1-c_2}{2} ((z+p)^2-(z-p)^2)\\
& = 2zp(1-c_2)
\end{align}
$$

and

$$
\begin{align}
\int_{z-p}^{z+p} (1-r^2)^{1/2} r dr &= -\frac{1}{2}\int_{1-(z-p)^2}^{1-(z+p)^2} u^{1/2} du \\
& = -\frac{1}{2}\frac{2}{3} \left.(1-r^2)^{3/2}\right|_{(z-p)}^{z+p}\\
& = -\frac{1}{3}\left[(1-(z+p)^2)^{3/2} - (1-(z-p)^2)^{3/2}\right].
\end{align}
$$

With this, 

$$
\begin{align}
I^*(z) &= \frac{1}{2zp}\left(2zp(1-c_2) - \frac{c_2}{3}\left[(1-(z+p)^2)^{3/2} - (1-(z-p)^2)^{3/2}\right]\right) \\
& = 1 - c_2 - \frac{c_2}{6zp}\left[(1-(z+p)^2)^{3/2} - (1-(z-p)^2)^{3/2}\right]
\end{align}
$$

The constant $\Omega$ which appears with $I^*(z)$ in the formula for $F$ is, from [Madel & Agol](http://adsabs.harvard.edu/abs/2002ApJ...580L.171M), just below Eq. (2), 

$$
\Omega = \sum_{n=0}^{4} \frac{c_n}{n+4},
$$

With linear darkening only, $c_0 = 1-c_2$, so 

$$
\Omega = \frac{1-c_2}{4} + \frac{c_2}{6} = \frac{3-c_2}{12}
$$
(note again that what [Poddaný et al (2010)](http://adsabs.harvard.edu/abs/2010NewA...15..297P) call $c_1$ is called $c_2$ by [Madel & Agol](http://adsabs.harvard.edu/abs/2002ApJ...580L.171M)).

## Outline of calculation

Given measurements of magnitude $m_i$ at times $t_i$ we want to find $A$, $B$, $C$, $t_0$, $D$ and $p$ that minimize the fit for $m(t_i)$ given in the first equation in this notebook. [Poddaný et al (2010)](http://adsabs.harvard.edu/abs/2010NewA...15..297P) hold the impact parameter $z_0$ (they call it $b$) fixed, using a value calculated from the orbital parameters of the exoplanet, and fix $c_1=0.5$.

1. Calculate $z(t_i)$; note this depends on the midpoint time $t_0$ and duration $D$.
2. Calculate $I^*(z)$; this depends directly on $p$ and $c_1$.
3. Calculate $F(z, p, c_1)$
4. Calculate $m(t_i; A, B, C, t_0, D, p)$, the model magnitude at time $t_i$; here the arguments after the semicolon are the fit parameters.
5. Minimize the difference between the model magnitudes and input magnitudes.
